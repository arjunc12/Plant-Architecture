```python
import numpy as np
from scipy.integrate import quad

np.random.seed(42)
n_tests = 10000

# Generate random test cases
x0_vals = np.random.uniform(0, 10, n_tests)
y0_vals = np.random.uniform(0, 10, n_tests)
p_vals  = np.random.uniform(0, 10, n_tests)
q_vals  = np.random.uniform(0, 10, n_tests)
G_vals  = np.random.uniform(0.01, 5, n_tests)  # avoid G=0

# Make sure p != x0 (would cause division by zero)
mask = np.abs(p_vals - x0_vals) > 0.01
x0_vals = x0_vals[mask]
y0_vals = y0_vals[mask]
p_vals  = p_vals[mask]
q_vals  = q_vals[mask]
G_vals  = G_vals[mask]

n_valid = len(x0_vals)
print(f"Testing {n_valid} random point pairs...")

max_diff = 0.0
n_failures = 0
tolerance = 1e-6

for i in range(n_valid):
    x0, y0 = x0_vals[i], y0_vals[i]
    p,  q  = p_vals[i],  q_vals[i]
    G      = G_vals[i]

    L_pos = pg.curve_length( G, x0, y0, p, q)
    L_neg = pg.curve_length(-G, x0, y0, p, q)
    diff  = abs(L_pos - L_neg)

    if diff > max_diff:
        max_diff = diff

    if diff > tolerance:
        n_failures += 1
        print(f"FAILURE at i={i}: x0={x0:.3f}, y0={y0:.3f}, "
              f"p={p:.3f}, q={q:.3f}, G={G:.3f} | "
              f"L(+G)={L_pos:.8f}, L(-G)={L_neg:.8f}, diff={diff:.2e}")

print(f"\nResults over {n_valid} tests:")
print(f"  Max |L(+G) - L(-G)|: {max_diff:.2e}")
print(f"  Failures (diff > {tolerance:.0e}): {n_failures}")
print(f"  {'THEOREM CONFIRMED' if n_failures == 0 else 'FAILURES DETECTED'}: "
      f"arc length is always equal for +G and -G")
```

```python
def curve_length_numerical(G, x0, y0, p, q):
    """Compute arc length via numerical integration, independent of pg.curve_length."""
    b, c = pg.calc_coeff(G, x0, y0, p, q)
    integrand = lambda x: np.sqrt(1 + (2*G*x + b)**2)
    length, _ = quad(integrand, min(x0, p), max(x0, p))
    return length

def parabola(x, G, b, c):
    return G*x**2 + b*x + c
```

```python
# -------------------------
# Test 1: calc_coeff boundary conditions
# For random (x0,y0), (p,q) and G, verify parabola passes through both points
# -------------------------
print("=" * 60)
print("TEST 1: calc_coeff boundary conditions")
print("=" * 60)

np.random.seed(42)
n_tests = 10000
tolerance = 1e-8
n_failures = 0
max_err = 0.0

for i in range(n_tests):
    x0 = np.random.uniform(-10, 10)
    y0 = np.random.uniform(-10, 10)
    p  = np.random.uniform(-10, 10)
    q  = np.random.uniform(-10, 10)
    G  = np.random.uniform(-5, 5)

    if abs(p - x0) < 0.01:
        continue

    b, c = pg.calc_coeff(G, x0, y0, p, q)

    # Check parabola passes through (x0, y0)
    y0_check = parabola(x0, G, b, c)
    err0 = abs(y0_check - y0)

    # Check parabola passes through (p, q)
    q_check = parabola(p, G, b, c)
    errp = abs(q_check - q)

    max_err = max(max_err, err0, errp)

    if err0 > tolerance or errp > tolerance:
        n_failures += 1
        print(f"FAILURE at i={i}: x0={x0:.3f}, y0={y0:.3f}, "
              f"p={p:.3f}, q={q:.3f}, G={G:.3f}")
        print(f"  y0 error={err0:.2e}, q error={errp:.2e}")

print(f"Tested {n_tests} random cases")
print(f"Max boundary error: {max_err:.2e}")
print(f"Failures: {n_failures}")
print(f"{'PASS' if n_failures == 0 else 'FAIL'}: calc_coeff correctly passes through both endpoints\n")

# -------------------------
# Test 2: arc length symmetry — numerical integration confirms L(+G) = L(-G)
# -------------------------
print("=" * 60)
print("TEST 2: Arc length symmetry L(+G) = L(-G) via numerical integration")
print("=" * 60)

n_failures = 0
max_diff = 0.0
tolerance = 1e-6

for i in range(n_tests):
    x0 = np.random.uniform(-5, 5)
    y0 = np.random.uniform(-5, 5)
    p  = np.random.uniform(-5, 5)
    q  = np.random.uniform(-5, 5)
    G  = np.random.uniform(0.01, 5)

    if abs(p - x0) < 0.01:
        continue

    L_pos = curve_length_numerical( G, x0, y0, p, q)
    L_neg = curve_length_numerical(-G, x0, y0, p, q)
    diff  = abs(L_pos - L_neg)
    max_diff = max(max_diff, diff)

    if diff > tolerance:
        n_failures += 1
        print(f"FAILURE at i={i}: x0={x0:.3f}, y0={y0:.3f}, "
              f"p={p:.3f}, q={q:.3f}, G={G:.3f}")
        print(f"  L(+G)={L_pos:.8f}, L(-G)={L_neg:.8f}, diff={diff:.2e}")

print(f"Tested {n_tests} random cases using scipy.integrate.quad")
print(f"Max |L(+G) - L(-G)|: {max_diff:.2e}")
print(f"Failures: {n_failures}")
print(f"{'PASS — THEOREM CONFIRMED' if n_failures == 0 else 'FAIL'}: "
      f"arc length is always equal for +G and -G\n")

# -------------------------
# Test 3: closed form vs numerical integration agree
# -------------------------
print("=" * 60)
print("TEST 3: Closed form curve_length vs numerical integration")
print("=" * 60)

n_failures = 0
max_diff = 0.0
tolerance = 1e-5  # slightly looser — numerical integration has its own error

for i in range(n_tests):
    x0 = np.random.uniform(-5, 5)
    y0 = np.random.uniform(-5, 5)
    p  = np.random.uniform(-5, 5)
    q  = np.random.uniform(-5, 5)
    G  = np.random.uniform(0.01, 5)

    if abs(p - x0) < 0.01:
        continue

    L_closed  = pg.curve_length(G, x0, y0, p, q)
    L_numeric = curve_length_numerical(G, x0, y0, p, q)
    diff = abs(L_closed - L_numeric)
    max_diff = max(max_diff, diff)

    if diff > tolerance:
        n_failures += 1
        print(f"FAILURE at i={i}: x0={x0:.3f}, y0={y0:.3f}, "
              f"p={p:.3f}, q={q:.3f}, G={G:.3f}")
        print(f"  closed={L_closed:.8f}, numerical={L_numeric:.8f}, diff={diff:.2e}")

print(f"Tested {n_tests} random cases")
print(f"Max |closed - numerical|: {max_diff:.2e}")
print(f"Failures (diff > {tolerance:.0e}): {n_failures}")
print(f"{'PASS' if n_failures == 0 else 'FAIL'}: closed form matches numerical integration")
```
