```python
import pandas as pd
```

```python
ga_d5_df = pd.read_csv('pimpi_GA_D5_Root_Nodes.csv', skipinitialspace=True)
ga_d5_df
```

```python
def reformat_root_name(root_name):
    root_name_items = root_name.split('_')
    if len(root_name_items) == 2:
        return root_name
    
    plate, condition, hormone, genotype = root_name.split('_')
    condition = condition.capitalize()
    replicate = plate.strip('pl')
    if not replicate.isnumeric():
        print(replicate)
    assert replicate.isnumeric()
    return '%s_%s_%s_%s' % (genotype, replicate, condition, hormone)
```

```python
ga_d5_df['root_name'] = ga_d5_df['root_name'].apply(reformat_root_name)
ga_d5_df
```

```python
ga_d5_df['image'] = '_set1_day5_' + ga_d5_df['image']
ga_d5_df
```

```python
ga_d5_df.to_csv('pimpi_GA_D5_Root_Nodes.csv', index=False)
```
