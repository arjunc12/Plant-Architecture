```python
import pandas as pd
import pareto_functions as pf
from constants import *
```

```python
arbor_name = '091_4_S_day5'
```

```python
pareto_front = pd.read_csv('%s/%s.csv' % (PARETO_FRONTS_DIR, arbor_name), skipinitialspace=True)
pareto_front
```

```python
tree_costs = pd.read_csv('%s/%s.csv' % (NULL_MODELS_DIR, arbor_name), skipinitialspace=True)
```

```python
tree_costs
```

```python
opt_wiring = pareto_front['wiring cost']
opt_delay = pareto_front['conduction delay']

for model, wiring, delay in zip(tree_costs['model'], tree_costs['wiring cost'], tree_costs['conduction delay']):
    print(model, wiring, delay)
    print(pf.pareto_dist(wiring, delay, opt_wiring, opt_delay))
```

```python
from scipy.spatial.distance import euclidean
for opt_w, opt_d in zip(opt_wiring, opt_delay):
    print(opt_w, opt_d)
    print('--->', euclidean((opt_w, opt_d), (15.080216, 14.620335)))
    print('--->', euclidean((opt_w, opt_d), (17.350800, 12.668243)))
```
